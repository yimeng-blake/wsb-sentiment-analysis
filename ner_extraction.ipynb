```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
from google.colab import files

ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", device=0)

file_path = "labeled_wsb.csv"
data = pd.read_csv(file_path)

text_column = "MergedColumn"

data = data.dropna(subset=[text_column])
data = data[data[text_column].str.strip() != ""]

def process_row(text):
    try:
        if pd.isna(text) or text.strip() == "":
            return False
        entities = ner_pipeline(text)

        reconstructed_entities = []
        current_entity = ""
        for entity in entities:
            if entity['word'].startswith("##"):
                current_entity += entity['word'][2:]
            else:
                if current_entity:
                    reconstructed_entities.append(current_entity)
                current_entity = entity['word']
        if current_entity:
            reconstructed_entities.append(current_entity)

        keywords = ["GME", "gme", "Gamestop", "Game stop", "gamestop", "GameStop"]
        return any(keyword.lower() in entity.lower() for entity in reconstructed_entities for keyword in keywords)
    except Exception as e:
        print(f"Error processing text: {text}\nError: {e}")
        return False

tqdm.pandas()
data["is_relevant"] = data[text_column].progress_apply(process_row)
relevant_data = data[data["is_relevant"]]
output_file = "filtered_dataset_gme.csv"
relevant_data.to_csv(output_file, index=False)

print(f"Filtered data saved to {output_file}")
files.download(output_file)
```
